```python
import os
import json
import re
import random
import time
import logging
import sys
from datetime import datetime

# Setup logging
logging.basicConfig(level=logging.INFO, format="%(levelname)s: %(message)s")

FILE_NAME = "tasks.json"

# Motivational messages
MOTIVATION = [
    "Great job! Keep going",
    "Consistency beats talent!",
    "You're building good habits",
    "Nice work today!"
]

def load_tasks():
    if os.path.exists(FILE_NAME):
        with open(FILE_NAME, "r") as f:
            return json.load(f)
    return []

def save_tasks(tasks):
    with open(FILE_NAME, "w") as f:
        json.dump(tasks, f, indent=2)

def validate_task(task):
    # Allow letters, numbers, spaces (basic validation)
    return re.fullmatch(r"[A-Za-z0-9 ]+", task)

def main():
    print("ðŸ“’ Study Task Logger")
    task = input("Enter a study task (or 'q' to quit): ")

    if task.lower() == "q":
        logging.info("User exited program")
        sys.exit()

    if not validate_task(task):
        logging.warning("Invalid task input")
        print("Task should only contain letters and numbers.")
        return

    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    tasks = load_tasks()
    tasks.append({
        "task": task,
        "time": timestamp
    })

    save_tasks(tasks)
    logging.info("Task saved successfully")

    time.sleep(0.5)  # Small delay
    print(random.choice(MOTIVATION))

if __name__ == "__main__":
    main()
```

```python
# ========================================
# Mini ML Project: Predict Scores
# ========================================

import pandas as pd
import numpy as np

# 1ï¸âƒ£ Load Data
df = pd.read_csv("study_data.csv")
X = df[["Study_Hours"]].values  # Features as NumPy array
y = df["Score"].values          # Target

# New sample to predict
new_sample = np.array([[4.5]])

# ========================================
# 2ï¸âƒ£ Scikit-learn: Linear Regression
# ========================================
from sklearn.linear_model import LinearRegression

sk_model = LinearRegression()
sk_model.fit(X, y)
sk_pred = sk_model.predict(new_sample)

print(f"Scikit-learn prediction: {sk_pred[0]:.2f}")

# ========================================
# 3ï¸âƒ£ TensorFlow/Keras: Neural Network
# ========================================
import tensorflow as tf
from tensorflow import keras

tf_model = keras.Sequential([
    keras.Input(shape=(1,)),          # Correct input layer
    keras.layers.Dense(10, activation='relu'),
    keras.layers.Dense(1)
])

tf_model.compile(optimizer='adam', loss='mean_squared_error')
tf_model.fit(X, y, epochs=100, verbose=0)
tf_pred = tf_model.predict(new_sample, verbose=0)

print(f"Keras/TensorFlow prediction: {tf_pred[0][0]:.2f}")

# ========================================
# 4ï¸âƒ£ PyTorch: Neural Network
# ========================================
import torch
import torch.nn as nn
import torch.optim as optim

X_torch = torch.tensor(X, dtype=torch.float32)
y_torch = torch.tensor(y, dtype=torch.float32).reshape(-1, 1)
new_sample_torch = torch.tensor(new_sample, dtype=torch.float32)

torch_model = nn.Sequential(
    nn.Linear(1, 10),
    nn.ReLU(),
    nn.Linear(10, 1)
)

criterion = nn.MSELoss()
optimizer = optim.Adam(torch_model.parameters(), lr=0.01)

# Training loop
for epoch in range(500):
    optimizer.zero_grad()
    output = torch_model(X_torch)
    loss = criterion(output, y_torch)
    loss.backward()
    optimizer.step()

with torch.no_grad():
    torch_pred = torch_model(new_sample_torch)

print(f"PyTorch prediction: {torch_pred.item():.2f}")

# ========================================
# 5ï¸âƒ£ XGBoost: High-Performance ML
# ========================================
import xgboost as xgb

xgb_model = xgb.XGBRegressor(n_estimators=100, random_state=42)
xgb_model.fit(X, y)
xgb_pred = xgb_model.predict(new_sample)

print(f"XGBoost prediction: {xgb_pred[0]:.2f}")

# ========================================
# âœ… Summary Table
# ========================================
print("\nPrediction Summary for Study Hours = 4.5")
print(f"{'Library':<20}{'Predicted Score'}")
print(f"{'-'*35}")
print(f"{'Scikit-learn':<20}{sk_pred[0]:.2f}")
print(f"{'Keras/TensorFlow':<20}{tf_pred[0][0]:.2f}")
print(f"{'PyTorch':<20}{torch_pred.item():.2f}")
print(f"{'XGBoost':<20}{xgb_pred[0]:.2f}")
```

```python
# ================================================
# Mini Project: Quotes Scraper API (Jupyter/Colab)
# ================================================

# 1ï¸âƒ£ Install dependencies (run once)
!pip install fastapi uvicorn nest_asyncio requests beautifulsoup4

# 2ï¸âƒ£ Imports
import requests
from bs4 import BeautifulSoup
from fastapi import FastAPI
import nest_asyncio
import uvicorn
import threading

# Allow nested event loops (for notebooks)
nest_asyncio.apply()

# ================================================
# 3ï¸âƒ£ Scrape Quotes from website
# ================================================
url = "http://quotes.toscrape.com/"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

quotes = []
for quote in soup.select(".quote"):
    text = quote.find("span", class_="text").get_text()
    author = quote.find("small", class_="author").get_text()
    quotes.append({"text": text, "author": author})

print("First 5 quotes:")
for q in quotes[:5]:
    print(q)

# ================================================
# 4ï¸âƒ£ Create FastAPI App
# ================================================
app = FastAPI(title="Quotes API")

@app.get("/")
def home():
    return {"message": "Welcome to the Quotes API!"}

@app.get("/quotes")
def get_all_quotes():
    return quotes

@app.get("/quotes/{author_name}")
def get_quotes_by_author(author_name: str):
    filtered = [q for q in quotes if q["author"].lower() == author_name.lower()]
    return filtered if filtered else {"message": "Author not found"}

# ================================================
# 5ï¸âƒ£ Run FastAPI App (Notebook-Friendly)
# ================================================
def run_api():
    config = uvicorn.Config(app=app, host="localhost", port=8010, log_level="info")
    server = uvicorn.Server(config=config)
    server.run()

# Start server in a separate thread
threading.Thread(target=run_api, daemon=True).start()
print("FastAPI server is running. Open http://localhost:8010/docs in your browser")
```
