```python
import pandas as pd

# Load the CSV file
df = pd.read_csv("data.csv")

print("Original Data:")
print(df)

# 1. Remove duplicate rows
df = df.drop_duplicates()

# 2. Handle missing values
# Fill missing age with the average age
df["age"] = df["age"].fillna(df["age"].mean())

# 3. Fix incorrect data
# Convert score column to numeric, forcing errors to NaN
df["score"] = pd.to_numeric(df["score"], errors="coerce")

# Replace missing/invalid scores with the average score
df["score"] = df["score"].fillna(df["score"].mean())

# 4. Ensure correct data types
df["age"] = df["age"].astype(int)
df["grade"] = df["grade"].astype(int)
df["score"] = df["score"].astype(int)

# 5. Standardize text (names)
df["name"] = df["name"].str.title()

print("\nCleaned Data:")
print(df)

# Save cleaned data to a new CSV file
df.to_csv("cleaned_data.csv", index=False)
```

```python
# Probability = Favorable outcomes / Total outcomes

# Example: Dice
favorable = 1   # getting a 6
total = 6

probability = favorable / total
print("Probability:", probability)
```

```python
# Given probabilities
P_A = 0.01        # Probability of disease
P_B_given_A = 0.9 # Test positive if disease
P_B = 0.1         # Test positive overall

# Bayes Theorem
P_A_given_B = (P_B_given_A * P_A) / P_B

print("Probability of disease given positive test:", P_A_given_B)
```

```python
# Simple Gradient Descent
learning_rate = 0.1
x = 10  # initial value

for i in range(10):
    gradient = 2 * x        # derivative of x^2
    x = x - learning_rate * gradient
    print(f"Step {i+1}: x = {x}")
```

```python
from sklearn.cluster import KMeans
import numpy as np

# Sample data: customers with (Age, Annual Spending)
X = np.array([[25, 2000], [40, 8000], [30, 2200], [45, 9000]])

# Create KMeans with 2 clusters
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X)

# Cluster labels for each point
print("Cluster labels:", kmeans.labels_)

# Centroids of the clusters
print("Centroids:\n", kmeans.cluster_centers_)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
import matplotlib.pyplot as plt
import numpy as np

# Sample data
X = np.array([[25, 2000], [40, 8000], [30, 2200], [45, 9000]])

# Compute linkage matrix
Z = linkage(X, method='ward')

# Plot dendrogram
plt.figure(figsize=(6, 4))
dendrogram(Z)
plt.show()

# Get clusters (cut dendrogram at distance 5000)
clusters = fcluster(Z, t=5000, criterion='distance')
print("Cluster labels:", clusters)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np

# Sample data: (Age, Annual Spending)
X = np.array([[25, 2000], [40, 8000], [30, 2200], [45, 9000]])

# Standardize features
X_std = StandardScaler().fit_transform(X)

# Apply PCA to reduce to 1 dimension
pca = PCA(n_components=1)
X_pca = pca.fit_transform(X_std)

print("Transformed data:\n", X_pca)
print("Explained variance ratio:", pca.explained_variance_ratio_)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
data = {
    'area': [800, 1000, 1200, 1500, 1800],
    'bedrooms': [1, 2, 2, 3, 3],
    'age': [20, 15, 10, 5, 2],
    'price': [3000000, 4000000, 5000000, 6500000, 7500000]
}

df = pd.DataFrame(data)
X = df[['area', 'bedrooms', 'age']]
y = df['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
model = LinearRegression()
model.fit(X_train, y_train)
prediction = model.predict([[1400, 3, 8]])
print("Predicted House Price:", prediction)
y_pred = model.predict(X_test)
print("MSE:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```
